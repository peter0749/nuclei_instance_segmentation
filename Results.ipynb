````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
loss = pd.read_csv('./run_.-tag-loss.csv')
val_loss = pd.read_csv('./run_.-tag-val_loss.csv')
mean_iou = pd.read_csv('./run_.-tag-mean_iou.csv')
val_mean_iou = pd.read_csv('./run_.-tag-val_mean_iou.csv')
mean_iou_marker = pd.read_csv('./run_.-tag-mean_iou_marker.csv')
val_mean_iou_marker = pd.read_csv('./run_.-tag-val_mean_iou_marker.csv')
loss = loss['Value']
val_loss = val_loss['Value']
mean_iou = mean_iou['Value']
val_mean_iou = val_mean_iou['Value']
mean_iou_marker = mean_iou_marker['Value']
val_mean_iou_marker = val_mean_iou_marker['Value']
````

````python
plt.plot(loss)
plt.plot(val_loss)
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','valid'], loc='upper right')
plt.show()
````

````python
plt.plot(mean_iou)
plt.plot(val_mean_iou)
plt.title('mean IoU')
plt.ylabel('mean IoU')
plt.xlabel('epoch')
plt.legend(['train','valid'], loc='upper right')
plt.show()
````

````python
plt.plot(mean_iou_marker)
plt.plot(val_mean_iou_marker)
plt.title('mean IoU (seed)')
plt.ylabel('mean IoU')
plt.xlabel('epoch')
plt.legend(['train','valid'], loc='upper right')
plt.show()
````

````python
print('validation loss: %.4f'%val_loss.iloc[-1])
print('validation mean IoU: %.4f'%val_mean_iou.iloc[-1])
print('validation mean IoU (seed): %.4f'%val_mean_iou_marker.iloc[-1])
````
